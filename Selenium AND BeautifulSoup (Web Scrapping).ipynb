```python
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome()

browser.get('http://www.yahoo.com')
assert 'Yahoo' in browser.title
print(browser.title)

elem = browser.find_element_by_name('p')  # Find the search box
elem.send_keys('seleniumhq' + Keys.RETURN)

browser.quit()
```

```python
master_listings = []

browser = webdriver.Chrome()

browser.get('https://www.glassdoor.com/index.htm')


search = browser.find_element_by_name('sc.keyword')  # Find the search box
search.send_keys("Data Science")
            

button = browser.find_element_by_id('HeroSearchButton')
button.click()

source = browser.page_source

soup = BeautifulSoup(source, 'html.parser')
joblistings = soup.find_all('li', class_="jl")    

for i in range(len(joblistings)):
    company = joblistings[i].find('div', class_="flexbox empLoc").get_text()
    position = joblistings[i].find('div', class_="flexbox").get_text()
    
    master_listings.append((position, company))

    
while True:
    elm = browser.find_element_by_class_name('next')
    if 'inactive' in elm.get_attribute('class'):
        break;
    elm.click()

    source = browser.page_source

    soup = BeautifulSoup(source, 'html.parser')
    joblistings = soup.find_all('li', class_="jl")    

    for i in range(len(joblistings)):
        company = joblistings[i].find('div', class_="flexbox empLoc").get_text()
        position = joblistings[i].find('div', class_="flexbox").get_text()

        master_listings.append((position, company))

    
browser.quit()
```
